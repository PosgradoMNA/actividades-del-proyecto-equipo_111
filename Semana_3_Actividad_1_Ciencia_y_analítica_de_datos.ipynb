# Parte 1: Fundamentos de bases de datos

**1. Fundamentos de bases de datos y para ciencia de datos.**

a.	La recopilación y limpieza de datos es uno de los pasos mas tardado dentro de la ciencia de datos, puede tomar hasta el 80% del tiempo del análisis. Es por esto que es de suma importancia las bases de datos. Una base de datos es una colección de información compuesta de una o mas listas (tablas) de datos que a su vez son organizadas por columnas, filas, y celdas. Contienen nombres de campo, campos, registros, celdas, datos, y objetos. Toda esta información se puede caracterizar como una recopilación de datos lógicamente relacionados con la intención de satisfacer las necesidades informacionales dentro de una organización. Dentro de organizaciones es de gran ayuda que esta información se encuentre en un repositorio central de datos compartidos y gestionados por un agente de control de una forma estandarizada y conveniente mediante un Database Management System (DBMS). Un DBMS es un sistema de software utilizado para crear, mantener, y proporcionar acceso controlado a usuarios a una base de datos. Algunas de las ventajas de un DBMS son reducir la redundancia de datos, mejorar la consistencia y calidad de los mismo, facilitar el intercambio de datos, y apoyar con una mejor toma de decisiones. Algunos puntos en contra de un DBMS es requerir de personal nuevo y especializado, el costo y complejidad de su instalación y conversión, y el posible conflicto organizacional.

**2.	Fundamentos de almacenes de datos (Data Warehouse) para ciencia de datos.**

a.	Un almacén de datos es un tipo de sistema para el manejo de información que está diseñado para dar soporte a la inteligencia de negocios, en especial a la analítica. Básicamente su propósito principal es el de permitir análisis sobre grandes cantidades de información histórica que proviene de muchas fuentes dentro de la organización. Debido a estas funciones y capacidades se le conoce como la fuente única de la verdad organizacional. Algunos de los beneficios son los siguientes: permiten el análisis de un tema o área en particular, crean consistencia entre diferentes tipos de información recopilada de varias fuentes, la información es estable, y permite revisas cambios sobre cierto tiempo. Un almacén de datos moderno permite agilizar el flujo de información. Esto significa que analistas, ingenieros, científicos de datos, y equipos de IT pueden utilizarlos mas eficientemente mientras buscan agregar innovación a la forma de trabajo para propulsar la organización hacia adelante, todo esto reduciendo el retraso y la complejidad. Un almacén de datos moderno debería buscar -mediante la exploración y durante la etapa de planeación- anticiparse a necesidades futuras. Su diseño debería permitir espacio para expandirse y evolucionar para mantener el paso con la evolución de las necesidades de los usuarios.


Segun la información contenida en el artículo de la empresa Oracle, especializada en manejos de bases de datos:  

"un almacen de datos(data warehouse) está destinado especificicamente al análisis de datos, tiene como características que puede manejar grandes cantidades de datos de varias fuentes, llevar un historico o que tienen cambios en el tiempo con el propósito de generar perspectivas basadas en el análisis, que sirven a las organización en el área de inteligencia de negocios."

# Parte 2: Selección y limpieza de los Datos en Python

```python
# Importar Pandas y Numpy

import pandas as pd
import numpy as np
```

```python
# Improtar la base de datos existente desde un link de internet

Data = pd.read_csv('https://raw.githubusercontent.com/PosgradoMNA/Actividades_Aprendizaje-/main/default%20of%20credit%20card%20clients.csv', index_col=0)
Data
```

```python
#Renombrado de columnas

Data = Data.rename(columns = {'X1': 'CantidadCredito', 
                          'X2': 'Genero', 
                          'X3': 'Escolaridad',
                          'X4': 'EstadoCivil',  
                          'X5': 'Edad',   
                          'X6': 'PagoACuentaSep',
                          'X7': 'PagoACuentaAgo', 
                          'X8': 'PagoACuentaJul', 
                          'X9': 'PagoACuentaJun',
                       'X10': 'PagoACuentaMay', 
                       'X11': 'PagoACuentaAbr', 
                       'X12': 'SaldoSep',
                       'X13': 'SaldoAgo', 
                       'X14': 'SaldoJul', 
                       'X15': 'SaldoJun',
                       'X16': 'SaldoMay', 
                       'X17': 'SaldoAbr', 
                       'X18': 'PagoPrevioSep',
                       'X19': 'PagoPrevioAgo', 
                       'X20': 'PagoPrevioJul', 
                       'X21': 'PagoPrevioJun', 
                       'X22': 'PagoPrevioMay', 
                       'X23': 'PagoPrevioAbr'})
Data
```

```python
# Verificar si hace fatla algun dato

Data.isnull().values.any()
```

```python
# Verificar en que columna hace falta el dato

Data.isnull().any()
```

```python
# Manera alternativa de validar si hace falta algun dato

Data.isna().values.any()
```

```python
# Visualizacion de la base de datos
Data
```

```python
# Eliminar filas con falantes, si eliminamos columnas confaltantes nos quedariamos con solo 1 columna. Las files se reducen de 30,000 a 29,958 lo cual significa una cantidad muy pequeña de perdida de información
Data.dropna(inplace = True)
Data
```

```python
# Confirmación de que ya no hay faltantes en las columnas

Data.isnull().any()
```

```python
# Copia de Data para Limpieza

Data2 = Data.copy()
```

```python
#Revision de columnas Genero, Escolaridad, EstadoCivil para confirmar si tienen la cantidad correcta de categorias como se nos dice en la descripcion de la base de datos

print(Data2.groupby(['Genero']).size()) # En Genero estamos bien, la descripcion comenta 2 categorias
print(Data2.groupby(['Escolaridad']).size()) # En Escolaridad tenemos un problema, la descripcion menciona 4 categorias y tenemos 6
print(Data2.groupby(['EstadoCivil']).size()) # En EstadoCivil tambien tenemos un problema con 4 categorias, la descripcion menciona 3 categorias

```

```python
#Como la cantidad de valores por categoria incorrecta es muy pequeña, se me hace logico incorporar las categorias incorrectas a la categoria correcta mas cercana. Es posible que los errores se hayan generado al momento de la entrada de datos

Data2['Escolaridad'].replace(0, 1, inplace=True)
Data2['Escolaridad'].replace(5, 4, inplace=True)
Data2['Escolaridad'].replace(6, 4, inplace=True)
Data2['EstadoCivil'].replace(0, 1, inplace=True)
```

```python
# Confirmacion que ahora todas las categorias concuerdan con lo mencionado en la descripcion

print(Data2.groupby(['Genero']).size())
print(Data2.groupby(['Escolaridad']).size())
print(Data2.groupby(['EstadoCivil']).size())
```
