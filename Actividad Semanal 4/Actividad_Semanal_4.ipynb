## Limpieza de datos realizada en semana 3

```python
# Importar Pandas y Numpy

import pandas as pd
import numpy as np
```

```python
# Importar la base de datos existente desde un link de internet

Data = pd.read_csv('https://raw.githubusercontent.com/PosgradoMNA/Actividades_Aprendizaje-/main/default%20of%20credit%20card%20clients.csv', index_col=0)
Data
```

```python
#Renombrado de columnas

Data = Data.rename(columns = {'X1': 'CantidadCredito', 
                          'X2': 'Genero', 
                          'X3': 'Escolaridad',
                          'X4': 'EstadoCivil',  
                          'X5': 'Edad',   
                          'X6': 'PagoACuentaSep',
                          'X7': 'PagoACuentaAgo', 
                          'X8': 'PagoACuentaJul', 
                          'X9': 'PagoACuentaJun',
                       'X10': 'PagoACuentaMay', 
                       'X11': 'PagoACuentaAbr', 
                       'X12': 'SaldoSep',
                       'X13': 'SaldoAgo', 
                       'X14': 'SaldoJul', 
                       'X15': 'SaldoJun',
                       'X16': 'SaldoMay', 
                       'X17': 'SaldoAbr', 
                       'X18': 'PagoPrevioSep',
                       'X19': 'PagoPrevioAgo', 
                       'X20': 'PagoPrevioJul', 
                       'X21': 'PagoPrevioJun', 
                       'X22': 'PagoPrevioMay', 
                       'X23': 'PagoPrevioAbr'})
```

```python
# Eliminar filas con falantes, si eliminamos columnas confaltantes nos quedariamos con solo 1 columna. Las filas se reducen de 30,000 a 29,958 lo cual significa una cantidad muy pequeña de perdida de información
Data.dropna(inplace = True)
Data
```

```python
# Confirmación de que ya no hay faltantes en las columnas

Data.isnull().any()
```

# Paso 1: Determine el número mínimo de componentes principales que representan la mayor parte de la variación en sus datos

```python
#Importar liberias que vamos a utilizar para aplicar PCA y analizar componentes
import matplotlib.pyplot as plt 
from sklearn.decomposition import PCA
from sklearn import preprocessing
```

```python
#Crear copia de la base de datos para trabajar sobre la copia
Data2=Data.copy()

Data2
```

```python
#Eliminar variables categoricas conocidas en el archivo, nos quedaran 14 columnas
Data3=Data2.drop(["Genero","Escolaridad","EstadoCivil","PagoACuentaSep","PagoACuentaAgo","PagoACuentaJul","PagoACuentaJun","PagoACuentaMay","PagoACuentaAbr","Y"], axis=1)
Data3
```

```python
#Estandarizar los datos
pcs = PCA()
pcs.fit(preprocessing.scale(Data3.iloc[:, 0:14].dropna(axis=0))) 
```

```python
#Intentamos encontrar cuantos componenetes son necesarios para llegar al 90% de la información. Para eso imprimimos la desviación estandar, la proporción de la varianzam y la propoción acumulada.

pcsSummary_df = pd.DataFrame({'Standard deviation': np.sqrt(pcs.explained_variance_),
                             'Proportion of variance': pcs.explained_variance_ratio_,
                             'Cumulative proportion': np.cumsum(pcs.explained_variance_ratio_)
                             })

pcsSummary_df = pcsSummary_df.transpose()
pcsSummary_df.columns = ['PC{}'.format(i) for i in range(1, len(pcsSummary_df.columns) + 1)]
pcsSummary_df.round(4) 
```

Basandonos en la proporción acumulada podemos ver que necesitamos 8 componentes principales que explican 90% de la información.

# Paso 2: Interprete cada componente principal en términos de las variables originales

```python
pcsComponents_df = pd.DataFrame(pcs.components_.transpose(), 
                                columns = pcsSummary_df.columns,
                                index = Data3.iloc[:, 0:14].columns
                                )
pcsComponents_df.iloc[:,:8]
```

```python
#Convertir pcsComponents en valores absolutos para hacer mas facil su revision
abspcsComponents=pd.DataFrame(data=pcsComponents_df.abs())
```

```python
#Analisis de las dos variables más importantes para el calculo de cada componente

print("Las dos variables más importantes para PCA 1 son: \n", abspcsComponents.PC1.nlargest(2),"\n")
print("Las dos variables más importantes para PCA 2 son: \n", abspcsComponents.PC2.nlargest(2),"\n")
print("Las dos variables más importantes para PCA 3 son: \n", abspcsComponents.PC3.nlargest(2),"\n")
print("Las dos variables más importantes para PCA 4 son: \n", abspcsComponents.PC4.nlargest(2),"\n")
print("Las dos variables más importantes para PCA 5 son: \n", abspcsComponents.PC5.nlargest(2),"\n")
print("Las dos variables más importantes para PCA 6 son: \n", abspcsComponents.PC6.nlargest(2),"\n")
print("Las dos variables más importantes para PCA 7 son: \n", abspcsComponents.PC7.nlargest(2),"\n")
print("Las dos variables más importantes para PCA 8 son: \n", abspcsComponents.PC8.nlargest(2),"\n")
```

```python
#Analisis de las dos variables menos importantes para el calculo de cada componente

print("Las dos variables menos importantes para PCA 1 son: \n", abspcsComponents.PC1.nsmallest(2),"\n")
print("Las dos variables menos importantes para PCA 2 son: \n", abspcsComponents.PC2.nsmallest(2),"\n")
print("Las dos variables menos importantes para PCA 3 son: \n", abspcsComponents.PC3.nsmallest(2),"\n")
print("Las dos variables menos importantes para PCA 4 son: \n", abspcsComponents.PC4.nsmallest(2),"\n")
print("Las dos variables menos importantes para PCA 5 son: \n", abspcsComponents.PC5.nsmallest(2),"\n")
print("Las dos variables menos importantes para PCA 6 son: \n", abspcsComponents.PC6.nsmallest(2),"\n")
print("Las dos variables menos importantes para PCA 7 son: \n", abspcsComponents.PC7.nsmallest(2),"\n")
print("Las dos variables menos importantes para PCA 8 son: \n", abspcsComponents.PC8.nsmallest(2),"\n")
```

**¿Cuándo se recomienda realizar un PCA y qué beneficios ofrece para Machine Learning?**

Se recomienda utilizar PCA cuando sabemos que tenemos un data set con variables correlacionadas. Esto es debido a que el PCA nos ayuda a transformar variables correlacionadas a un conjunto de valores de variables no correlacionadas. En caso de que sepamos por certeza de antemano que nuestras variables no se encuentran correlacionadas entonces el PCA no debería ser utilizando. El PCA ofrece beneficios para Machine Learning ya que es una herramienta para identificar la información de nuestro dataset mas importante y representativa sin perder mucha de la información en la misma. Esto ayuda a que los recursos necesarios para el análisis de la información sean menor y tome menos tiempo.
